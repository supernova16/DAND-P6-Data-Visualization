```python
import csv
from bs4 import BeautifulSoup 
import requests


#url = 'http://applestorefronts.appspot.com/all/new/'
#html = requests.get(url).text

soup = BeautifulSoup(open('applestorefronts_all_new.html'),'lxml')

store_info = []

all_store = soup.find_all('div', class_ = 'row')[0]#find_all()返回的结果是一个可迭代对象，其实只有一个元素,所以取第一个就可以了


for store in all_store.find_all(class_ = 'ribbon'):
    num = store.find(class_ = 'anchor').string.split(' ')[0][1:]
    name = store.find(class_ = 'anchor').string.split(' ')[1:]
    name = ' '.join(name) #将列表变为字符串，列表项间用空格链接
    rnum = store.find(class_ = 'rnum').string
    if len(store.find_all('p')[1].get_text().split(',')) == 3:
        city = store.find_all('p')[1].get_text().split(',')[0]
        country = store.find_all('p')[1].get_text().split(',')[2]
    else:
        city = store.find_all('p')[1].get_text().split(',')[0]
        country = store.find_all('p')[1].get_text().split(',')[1]
    
    date = store.find_all('p')[2].get_text().split(': ')[1]
    
    store_info.append([num, name, rnum, city, country, date])



i = 0
for store in all_store.find_all(class_ = 'photo'):
    url = store.a.attrs['href']
    photo = store.img.attrs['data-href']
    store_info[i].append(url)
    store_info[i].append(photo)
    
    """
    html = requests.get(url).text
    soup2 = BeautifulSoup(html, 'lxml')
    
    
    addr1 = soup2.find_all(class_ = 'store-street typography-body')[:2][0].string
    addr2 = soup2.find_all(class_ = 'store-street typography-body')[:2][1].string
    if addr2 != None:
        addr = addr1 + ' ' + addr2
    else:
        addr = addr1
    
    #store_info[i].append(addr)
    
    tel = soup2.find_all(class_ = 'store-phone typography-body')[0].string
    #store_info[i].append(tel)
    
    long = soup2.find_all(class_ = 'more typography-body')[1].attrs['href'].split('&')[4][5:]
    lat = soup2.find_all(class_ = 'more typography-body')[1].attrs['href'].split('&')[3][4:]
    print(lat)
    """
    i += 1
    

with open('apple_store_list.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Num','Name','ID','City','Country','Date','URL','Photo','Address','Tel'])
    for row in store_info:
        writer.writerow(row)
    
```
